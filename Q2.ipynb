```python
numerical_columns = ['Monthly_Inhand_Salary', 'Num_of_Delayed_Payment', 'Annual_Income',
                     'Num_Credit_Inquiries', 'Amount_invested_monthly', 'Monthly_Balance']
```

```python
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.head()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].apply(lambda x: None if x < 0 else x)
```

```python
# Impute missing values
df['Name'].fillna(df['Name'].mode()[0], inplace=True)
df['Monthly_Inhand_Salary'].fillna(df['Monthly_Inhand_Salary'].median(), inplace=True)
df['Type_of_Loan'].fillna('Not Specified', inplace=True)
df['Num_of_Delayed_Payment'].fillna(df['Num_of_Delayed_Payment'].median(), inplace=True)
df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].median(), inplace=True)
df['Credit_History_Age'].fillna(df['Credit_History_Age'].mode()[0], inplace=True)
df['Amount_invested_monthly'].fillna(df['Amount_invested_monthly'].median(), inplace=True)
df['Monthly_Balance'].fillna(df['Monthly_Balance'].median(), inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Annual_Income', y='Monthly_Balance', data=df)
plt.title('Annual Income vs Monthly Balance')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# List of columns to plot (categorical for count plots and pie charts, numerical for box plots)
categorical_columns = ['Occupation', 'Credit_Mix', 'Type_of_Loan', 'Payment_of_Min_Amount']
numerical_columns = ['Annual_Income', 'Monthly_Balance', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries','Age']

# Filter categorical columns with a manageable number of categories
categorical_columns = [col for col in categorical_columns if df[col].nunique() <= 10]

# Function to plot count plot
def plot_countplot(ax, df, col):
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=ax)
    ax.set_title(f'Count Plot of {col}')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Function to plot box plot for numerical columns
def plot_boxplot(ax, df, col):
    sampled_df = df[col].dropna()
    if len(sampled_df) > 5000:  # Sample if data is too large
        sampled_df = sampled_df.sample(5000)
    sns.boxplot(x=sampled_df, ax=ax)
    ax.set_title(f'Box Plot of {col}')
```

```python
# Create grids for Count Plots and Pie Charts
fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=2, figsize=(15, 5 * len(categorical_columns)))
for i, col in enumerate(categorical_columns):
    plot_countplot(axes[i, 0], df, col)
    df[col].value_counts().plot.pie(ax=axes[i, 1], autopct='%1.1f%%', startangle=90, cmap='viridis')
    axes[i, 1].set_title(f'Pie Chart of {col}')
    axes[i, 1].set_ylabel('')

plt.tight_layout()
plt.show()

# Create grids for Box Plots
fig, axes = plt.subplots(nrows=1, ncols=len(numerical_columns), figsize=(5 * len(numerical_columns), 5))
for i, col in enumerate(numerical_columns):
    plot_boxplot(axes[i], df, col)

plt.tight_layout()
plt.show()
```

```python
def handle_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_value = column.median()
    # Replace outliers with the median value
    return column.apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

# Apply the outlier handling function to each numerical column
for col in numerical_columns:
    df[col] = handle_outliers(df[col])
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```
